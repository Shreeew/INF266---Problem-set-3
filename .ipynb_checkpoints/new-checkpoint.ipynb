```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import random
```

```python
def create_bins(num_bins=20):
    # For MountainCar-v0: position in [-1.2, 0.6], velocity in [-0.07, 0.07]
    pos_bins = np.linspace(-1.2, 0.6, num_bins - 1)
    vel_bins = np.linspace(-0.07, 0.07, num_bins - 1)
    return pos_bins, vel_bins

def discretize_state(state, bins):
    # Ensure bins is a tuple (pos_bins, vel_bins), not a dict
    if isinstance(bins, dict):
        raise ValueError("The 'bins' variable is a dict. Please ensure you call create_bins() to obtain a tuple of (pos_bins, vel_bins).")
    pos_bins, vel_bins = bins
    pos, vel = state
    pos_idx = np.digitize(pos, pos_bins)
    vel_idx = np.digitize(vel, vel_bins)
    return (pos_idx, vel_idx)
```

```python
def random_policy(state, env):
    return env.action_space.sample()

def straight_policy(state, env):
    # For MountainCar-v0, action 2 (accelerate right) is commonly used.
    return 2

def greedy_policy(state, Q, env, epsilon=0.0):
    # -greedy: with probability epsilon choose a random action.
    if random.random() < epsilon:
        return env.action_space.sample()
    else:
        q_values = Q[state]
        return int(np.argmax(q_values))
```

```python
def run_episode(env, policy_fn, bins, Q=None, max_steps=200, epsilon=0.0):
    state = env.reset()
    disc_state = discretize_state(state, bins)
    episode = []
    for t in range(max_steps):
        # If Q is provided, assume the policy uses it (for example, greedy_policy)
        if Q is None:
            action = policy_fn(state, env)
        else:
            action = policy_fn(disc_state, Q, env, epsilon)
        next_state, reward, done, _ = env.step(action)
        episode.append((disc_state, action, reward))
        disc_state = discretize_state(next_state, bins)
        state = next_state
        if done:
            break
    return episode
```

```python
def mc_control(env, num_episodes=5000, gamma=1.0, epsilon=0.1, bins=None):
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    returns_sum = defaultdict(float)
    returns_count = defaultdict(float)
    if bins is None:
        bins = create_bins()
        
    final_episode_trajectory = []

    for i_episode in range(1, num_episodes + 1):
        # Generate an episode using the current Q and -greedy policy
        episode = run_episode(env, greedy_policy, bins, Q, max_steps=200, epsilon=epsilon)
        final_episode_trajectory = episode  # Store the final episode for visualization

        # First-visit MC update
        sa_in_episode = set((x[0], x[1]) for x in episode)
        for state, action in sa_in_episode:
            first_occurrence = next(i for i, x in enumerate(episode) if x[0] == state and x[1] == action)
            G = sum(x[2] * (gamma ** i) for i, x in enumerate(episode[first_occurrence:]))
            returns_sum[(state, action)] += G
            returns_count[(state, action)] += 1.0
            Q[state][action] = returns_sum[(state, action)] / returns_count[(state, action)]
        
        # Optional: Decay epsilon over episodes for more exploitation later
        # epsilon = max(0.01, epsilon * 0.999)

    return Q, final_episode_trajectory, bins
```

```python
def mc_control(env, num_episodes=5000, gamma=1.0, epsilon=0.1, bins=None):
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    returns_sum = defaultdict(float)
    returns_count = defaultdict(float)
    if bins is None:
        bins = create_bins()
        
    final_episode_trajectory = []

    for i_episode in range(1, num_episodes + 1):
        # Generate an episode using the current Q and -greedy policy
        episode = run_episode(env, greedy_policy, bins, Q, max_steps=200, epsilon=epsilon)
        final_episode_trajectory = episode  # Store the final episode for visualization

        # First-visit MC update
        sa_in_episode = set((x[0], x[1]) for x in episode)
        for state, action in sa_in_episode:
            first_occurrence = next(i for i, x in enumerate(episode) if x[0] == state and x[1] == action)
            G = sum(x[2] * (gamma ** i) for i, x in enumerate(episode[first_occurrence:]))
            returns_sum[(state, action)] += G
            returns_count[(state, action)] += 1.0
            Q[state][action] = returns_sum[(state, action)] / returns_count[(state, action)]
        
        # Optional: Decay epsilon over episodes for more exploitation later
        # epsilon = max(0.01, epsilon * 0.999)

    return Q, final_episode_trajectory, bins
```

```python
def plot_trajectory(trajectory, title="Trajectory on Mountain Map"):
    # Extract discretized position and velocity indices for plotting
    positions = [s[0] for (s, a, r) in trajectory]
    velocities = [s[1] for (s, a, r) in trajectory]
    
    plt.figure(figsize=(8, 6))
    plt.plot(positions, velocities, marker='o', linestyle='-', color='b')
    plt.title(title)
    plt.xlabel("Discretized Position")
    plt.ylabel("Discretized Velocity")
    plt.grid(True)
    plt.show()
```

```python
# Initialize the MountainCar environment
bins = create_bins(num_bins=20)

# Testing initial policies: Random, Straight, and Greedy
initial_policies = {
    "Random": random_policy,
    "Straight": straight_policy,
    # For the Greedy policy here, Q-values are initialized to zero so it acts deterministically.
    "Greedy": lambda state, env: greedy_policy(discretize_state(state, bins), 
                                                 defaultdict(lambda: np.zeros(env.action_space.n)), 
                                                 env, epsilon=0.0)
}

for name, policy_fn in initial_policies.items():
    episode = run_episode(env, policy_fn, bins, max_steps=200)
    plot_trajectory(episode, title=f"Trajectory using {name} Policy")

# -------------------------------
# Perform Monte Carlo Control for Policy Improvement
# -------------------------------
print("Starting Monte Carlo Control...")
Q, final_episode, bins = mc_control(env, num_episodes=5000, gamma=1.0, epsilon=0.1, bins=bins)

# Define an improved policy derived from the learned Q-values
def improved_policy_fn(state, env, epsilon=0.0):
    # 'state' here is assumed to be discretized already.
    return int(np.argmax(Q[state]))

# Run an episode with the improved policy
episode_improved = run_episode(env, lambda s, env: improved_policy_fn(s, env), bins, max_steps=200)
plot_trajectory(episode_improved, title="Trajectory using Improved MC-Control Policy")

env.close()
```
