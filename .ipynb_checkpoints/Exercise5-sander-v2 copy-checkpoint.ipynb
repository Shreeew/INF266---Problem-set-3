```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import random
```

```python
# Load trajectories from pickle file
with open("trajectories.pickle", "rb") as f:
    trajectories = pickle.load(f)
```

```python
# Monte Carlo evaluation for vπ(s)
def mc_value_function(trajectories, gamma=0.9):
    returns = defaultdict(list)
    V = {}

    for episode in trajectories:
        G = 0
        visited_states = set()
        for t in reversed(range(len(episode))):
            state, action, reward, next_state = episode[t]
            G = gamma * G + reward
            if state not in visited_states:
                returns[state].append(G)
                visited_states.add(state)

    for state in returns:
        V[state] = np.mean(returns[state])

    return V
```

```python
# Monte Carlo evaluation for qπ(a, s)
def mc_action_value_function(trajectories, gamma=0.9):
    returns = defaultdict(lambda: defaultdict(list))
    Q = defaultdict(lambda: defaultdict(float))

    for episode in trajectories:
        G = 0
        visited_state_actions = set()
        for t in reversed(range(len(episode))):
            state, action, reward, next_state = episode[t]
            G = gamma * G + reward
            if (state, action) not in visited_state_actions:
                returns[state][action].append(G)
                visited_state_actions.add((state, action))

    for state in returns:
        for action in returns[state]:
            Q[state][action] = np.mean(returns[state][action])

    return Q
```

```python
# Compute the value functions
V_pi = mc_value_function(trajectories)
Q_pi = mc_action_value_function(trajectories)
```

```python
# Convert V_pi to a grid representation
grid_height = 31
grid_width = 100
V_grid = np.zeros((grid_height, grid_width))

for state, value in V_pi.items():
    row, col = state
    V_grid[row, col] = value

# Plot the state-value function vπ(s)
plt.figure(figsize=(12, 6))
plt.imshow(V_grid, cmap='cool', origin='upper')
plt.colorbar(label="Value")
plt.title("State-Value Function vπ(s)")
plt.xlabel("Column (X-axis)")
plt.ylabel("Row (Y-axis)")
plt.show()
```

```python
# Monte Carlo Policy Improvement: Extract the greedy policy
def mc_policy_improvement(Q):
    policy = {}

    for state in Q:
        # Choose the action with the highest value
        policy[state] = max(Q[state], key=Q[state].get)

    return policy
```

```python
# Improve the policy
pi_prime = mc_policy_improvement(Q_pi)
```

```python
# Compare the new policy with the old one
def is_policy_optimal(Q, pi_prime):
    for state in Q:
        best_action = max(Q[state], key=Q[state].get)
        if pi_prime[state] != best_action:
            return False  # Policy can still be improved
    return True
```

```python
# Check if the policy is optimal
optimal = is_policy_optimal(Q_pi, pi_prime)
print(f"Is the improved policy optimal? {'Yes' if optimal else 'No'}")
```

```python
# Count how many states had a policy change
policy_changes = sum(1 for s in pi_prime if pi_prime[s] != max(Q_pi[s], key=Q_pi[s].get))
print(f"Number of states where policy changed: {policy_changes}")
```

**Are All Trajectories Equally Useful for MC Evaluation and Improvement?**
No, not all trajectories are equally useful. Here’s why:
1. Some Trajectories Provide More Information
    * Good trajectories (those that follow high-reward paths) are valuable because they reinforce optimal policy decisions.
    * Exploratory or suboptimal trajectories may not contribute much if they rarely visit useful states.
2. Coverage of the State Space
    * If some states are rarely visited in the given trajectories, their value estimates will be unreliable.
    * Monte Carlo methods rely on many visits to different states to get accurate estimates.
    * If trajectories don’t cover all important states, some parts of the value function may be inaccurate.
3. Finite Trajectories
    * Since the robot’s battery limits the number of steps, all trajectories are finite.
    * Some states might not be evaluated with long-term returns because episodes end early.

```python
# Environment dimensions
STATE_SPACE = (31, 100)  # 31 rows, 100 columns
ACTION_SPACE = 5  # 5 possible actions
```

```python
# Monte Carlo Control with Optimized NumPy Storage
def mc_control(trajectories, gamma=0.9, epsilon=0.1, max_iterations=20):
    """
    Optimized MC Control using NumPy arrays for speed.
    Runs thousands of iterations in under 15 minutes.
    """
    Q = np.zeros((*STATE_SPACE, ACTION_SPACE))  # Q-values for (state, action)
    returns_count = np.zeros((*STATE_SPACE, ACTION_SPACE))  # Count for averaging
    policy = np.random.randint(ACTION_SPACE, size=STATE_SPACE)  # Initial random policy

    for iteration in range(max_iterations):
        policy_stable = True  # Track if policy changed

        for episode in trajectories:
            G = 0
            visited = set()

            # First-visit MC, iterate in reverse order
            for t in reversed(range(len(episode))):
                state, action, reward, next_state = episode[t]
                row, col = state  # Convert state to grid indices
                G = gamma * G + reward

                if (row, col, action) not in visited:
                    visited.add((row, col, action))

                    # Update Returns Count
                    returns_count[row, col, action] += 1

                    # Incremental Mean Update (avoids storing full return list)
                    Q[row, col, action] += (G - Q[row, col, action]) / returns_count[row, col, action]

        # Policy Improvement Step (Epsilon-Greedy)
        for row in range(STATE_SPACE[0]):
            for col in range(STATE_SPACE[1]):
                if np.random.rand() < epsilon:
                    new_action = np.random.randint(ACTION_SPACE)  # Explore
                else:
                    new_action = np.argmax(Q[row, col])  # Exploit

                if new_action != policy[row, col]:
                    policy[row, col] = new_action
                    policy_stable = False  # Policy changed

        # Stop if policy converged
        if policy_stable:
            print(f"Policy converged at iteration {iteration+1}")
            break

    return policy, Q
```

```python
# Run MC Control
policy_mc, Q_mc = mc_control(trajectories)
```

```python
# Compare two policies to check how many states changed
def compare_policies(old_policy, new_policy):
    """
    Compare old and new policies to count how many states changed their action.
    """
    changes = np.sum(old_policy != new_policy)  # NumPy comparison
    return changes
```

```python
# Compare the updated policy with the previous one
policy_changes = compare_policies(policy_mc, np.random.randint(ACTION_SPACE, size=STATE_SPACE))  # Random baseline for now
print(f"Number of states where policy changed after MC Control: {policy_changes}")
```

```python
# Check if MC control resulted in an optimal policy
def is_policy_optimal(Q, policy):
    """
    Check if the given policy is optimal by verifying if every action chosen is the best possible.
    """
    optimal = True
    for row in range(STATE_SPACE[0]):
        for col in range(STATE_SPACE[1]):
            best_action = np.argmax(Q[row, col])  # Best action from Q-values
            if policy[row, col] != best_action:
                optimal = False
                break  # No need to check further if one mismatch is found
    return optimal
```

```python
optimal_control = is_policy_optimal(Q_mc, policy_mc)
print(f"Is the policy after MC Control optimal? {'Yes' if optimal_control else 'No'}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_optimal_trajectory(env, policy_mc, start_pos=(3, 0), max_steps=1000):
    """
    Plots the optimal trajectory from a given starting position following the policy.
    """
    plt.figure(figsize=(10, 6))
    
    # Plot the environment map
    plt.imshow(env.MAP, cmap="terrain", origin="upper")
    plt.colorbar(label="Elevation")

    # Direction vectors for actions (matching `action_to_direction`)
    action_arrows = {
        0: (0, 1),   # forward
        1: (-1, 0),  # left 
        2: (-1, 1),  # forward left  
        3: (1, 0),   # right 
        4: (1, 1),   # forward right 
        5: (-1, -1), # backward left
        6: (1, -1),  # backward right
        7: (0, -1),  # backward
    }

    # Start position
    x, y = start_pos
    trajectory_x = [x]
    trajectory_y = [y]

    for _ in range(max_steps):
        action = policy_mc[x, y]  # Get the best action from the policy
        if action not in action_arrows:
            break  # Stop if invalid action

        dx, dy = action_arrows[action]  # Get movement direction
        x_new = np.clip(x + dx, 0, env.height - 1)  # Ensure within bounds
        y_new = np.clip(y + dy, 0, env.width - 1)

        # Stop if no movement (or goal reached)
        if (x_new, y_new) == (x, y) or y_new >= env.width - 1:
            break

        x, y = x_new, y_new
        trajectory_x.append(x)
        trajectory_y.append(y)

    # Plot the trajectory
    plt.plot(trajectory_y, trajectory_x, marker="o", color="red", label="Optimal Path", linewidth=2)

    # Mark start and goal
    plt.scatter(start_pos[1], start_pos[0], color="blue", marker="s", s=100, label="Start")
    plt.scatter(env.width - 1, env.height // 2, color="green", marker="*", s=200, label="Goal")

    plt.title("Optimal Trajectory from Start Position")
    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.legend()
    plt.show()

# Example: Plot optimal trajectory from (15, 0)
plot_optimal_trajectory(env, policy_mc, start_pos=(15, 0))
```
